```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4])
plt.ylabel('some numbers')
plt.show()
#plot() 으로 그래프 그릴 수 있음
#하나의 리스트로 값을 줄 경우 y 값으로 간주
#x값을 자동으로 0 부터 생성함
```

```python
#x,y 축 값을 가지는 그래프
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[1,4,9,16]) 
plt.ylabel('cnt')
plt.xlabel('time')
plt.show()
```

```python
#스타일 정하기
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[1,4,9,16],'ro') #세번째 인자로 포맷 스트링(색, 선종류, MATLAB 의 format string 따름)
plt.axis([0,6,0,20]) #x,y 축의 범위 조정
plt.show()
```

```python
#numpy 배열을 이용한 그래프 그리기
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0.,5.,0.2)

plt.plot(t,t,'r--', t,t**2,'bs',t,t**3,'g^')
plt.show()
```

```python
#키워드 문자열로 그래프 그리기
import matplotlib.pyplot as plt
import numpy as np

data = {'a': np.arange(25),
       'c': np.random.randint(0, 25, 25),
       'd':np.random.randn(25)}
data['b'] = data['a'] +10*np.random.randn(25)
data['d'] = np.abs(data['d'])*100

plt.scatter('a','b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
#범주화된 변수들을 이용한 그래프 그리기
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9,3)) # 가로 9인치, 세로 3인치

plt.subplot(131) #1: 행 개수, 3:열 개수, 1: 인덱스
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle("Categorical Plotting")
plt.show()
```

```python
#여러 그림과 축으로 그래프 그리기
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
#그림과 축을 변경하며 그래프 그리기 
import matplotlib.pyplot as plt
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])

plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title
```
